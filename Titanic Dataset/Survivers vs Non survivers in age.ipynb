```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df.head(1)
```

```python
pd.cut(df["Age"], bins = np.arange(0,110, 10)).sort_values()
```

```python
age_bin = pd.cut(df["Age"], bins = np.arange(0,110, 10)).sort_values()
```

```python
((df.groupby(age_bin)["Survived"].mean().round(2))*100).fillna(0)
```

```python
df.loc[df["Survived"] == 0, "Age"].fillna(df["Age"].mean())
```

```python
not_survived_age = df.loc[df["Survived"] == 0, "Age"].fillna(df["Age"].mean())
```

```python
pd.cut(not_survived_age, bins = np.arange(0,110, 10)).sort_values()
```
