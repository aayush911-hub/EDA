```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df.head(1)
```

```python
df.groupby(["Pclass"])["Survived"].count()
```

```python
survival_class = (df.groupby(["Pclass"])["Survived"].mean().round(2))*100
```

```python
survival_class
```

```python
survival_class.index
```

```python
survival_class.values
```

```python
plt.bar(survival_class.index, survival_class.values, color = ["Cyan", "Green", "Red"])
plt.xticks([1, 2, 3])
plt.xlabel("Classes")
plt.ylabel("Percentage% of Survival")
plt.title("Survival By Class")
plt.show()
```
