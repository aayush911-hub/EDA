```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df.head(1)
```

```python
df.groupby(["Sex"])["Survived"].count()
```

```python
(df.groupby(["Sex"])["Survived"].mean().round(2))*100
```

```python
survival_by_gender = (df.groupby(["Sex"])["Survived"].mean().round(2))*100
```

```python
plt.bar(survival_by_gender.index, survival_by_gender.values, color = ["violet", "Cyan"])
plt.xlabel("Gender")
plt.ylabel("Percentage% of Survival")
plt.title("Survival By Gender")
plt.show()
```
